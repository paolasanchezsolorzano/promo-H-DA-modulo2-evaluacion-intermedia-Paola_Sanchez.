```python

def llamar_senderos(url_senderos):

    url_senderos = "https://es.wikipedia.org/wiki/Sendero_de_Gran_Recorrido"
     # hacemos la request a la página de la que queremos sacar la info
    res_senderos = requests.get(url_senderos)

    # Verificamos si la petición fue exitosa
    print(f"La respuesta de la petición es:", res_senderos.status_code)

    # Creamos el objeto BeautifulSoup para acceder al contenido solicitado
    sopa_senderos = BeautifulSoup(res_senderos.content, "html.parser")

    # Buscamos todas las tablas en la página web
    tablas = sopa_senderos.findAll("table")
    print("Las tablas son un total de:", len(tablas))

    # La tabla que nos interesa es la primera (índice 0)
    mi_tabla = tablas[0]

    # Extraemos los encabezados de la tabla
    lista_de_encabezados = mi_tabla.findAll('th')
    print("\n.................................\n")  

    encabezados_senderos = [columna.text for columna in lista_de_encabezados]  
    
    encabezados_senderos= [ i.rstrip() for i in encabezados_senderos]
    print(encabezados_senderos[:-1])
    
    
    # Sacamos todas las filas de la tabla usando el método .find_all()
    filas_senderos = mi_tabla.find_all("tr")

    # Empezamos a iterar desde la segunda fila, ya que la primera contiene el encabezado
    resultados_filas= [] #creo dicionario
     
    for fila in filas_senderos[1:]:
        resultados_filas.append(fila.text.split("\n")[1:4])
    for indice_lista, lista in enumerate(resultados_filas):
        for indice_elemento, elemento in enumerate(lista):
             return  encabezados_senderos,resultados_filas
```

```python


senderos= llamar_senderos(url_senderos)

df_senderos = pd.DataFrame(resultados_filas,columns= encabezados_senderos[:-1])
df_senderos.head()
```
